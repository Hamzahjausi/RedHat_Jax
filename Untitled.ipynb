```python
from methods import forward_setup ,backward_setup
from calc import fast_calc
import jax.numpy as jnp
import matplotlib.pyplot as plt 
```

```python
A = jnp.array ([
    [0,1],
    [-1,0]
], dtype=jnp.float64)
ini = jnp.array ([1,0], dtype=jnp.float64)
t = 2 * jnp.pi
dt = 0.1 
```

```python
n_steps_F , U_matrix_F= forward_setup(A , dt , t)
n_steps_B ,U_matrix_B = backward_setup(A,dt,t)
U_F0,U_F1 = fast_calc(U_matrix_F , ini , n_steps_F).T
U_B0,U_B1 = fast_calc(U_matrix_B , ini , n_steps_B).T
plt.plot(U_F0, U_F1, marker='o', linestyle='-', color='b')
plt.plot(U_B0, U_B1, marker='o', linestyle='-', color='r')
```
